# Data Overview (From PostgreSQL):


```python
# db_table_data_overview.py

# Import necessary libraries for data overview
from sqlalchemy import create_engine, inspect
import pandas as pd

# --------------------------------------------
# DATABASE CONNECTION
# --------------------------------------------
DB_URI = "postgresql+psycopg2://postgres:Sansen27%40@127.0.0.1:5432/attendance_db"
engine = create_engine(DB_URI)
inspector = inspect(engine)

# --------------------------------------------
# LIST TABLES
# --------------------------------------------
tables = inspector.get_table_names()
print("\n=== TABLES IN DATABASE ===")
for t in tables:
    print("-", t)
print("\n")

# --------------------------------------------
# DISPLAY TABLE DETAILS AND DATA
# --------------------------------------------
for table_name in tables:
    print(f"=== TABLE: {table_name} ===")

    # Column details
    columns = inspector.get_columns(table_name)
    print("Attributes (Columns):")
    for col in columns:
        print(f"  - {col['name']} | {col['type']} | nullable: {col['nullable']}")
    print()

    # Fetch and display all rows
    df = pd.read_sql_table(table_name, con=engine)
    print("All Rows:")
    if df.empty:
        print("  (No rows found)")
    else:
        print(df.to_string(index=False))

    # Show shape of the data
    print(f"\nShape: {df.shape[0]} rows × {df.shape[1]} columns")
    print("=" * 90 + "\n")
```
